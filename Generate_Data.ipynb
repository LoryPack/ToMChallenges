## Create Story template

```python
class data_smarties:
  def __init__(self, name, obj, location, place, question):
    self.name = name ##[A,B]
    self.obj = obj ## [O1,O2]
    self.location = location ## L1
    self.place = place ##P1
    self.Q = question ##['reality', 'belief', '1stA', '1stB', '2ndA', '2ndB']
  def premise(self):
    sent1 = "{} found a {} in the {}.".format(self.name[0], self.place, self.location)
    sent2 = "The label on the {} says {}.".format(self.place, self.obj[0])
    sent3 = "{} couldnâ€™t see what was inside the {}.".format(self.name[0], self.place)
    sent4 = "{} opened the {} and found a {}.".format(self.name[0], self.place, self.obj[1])
    sent5 = "There is no {} in the {}.".format(self.obj[0], self.place)
    sent6 = "{} closed the {} and put it back.".format(self.name[0], self.place)
    sent7 = "{} entered the {} and saw the {}.".format(self.name[1], self.location, self.place)
    return [sent1, sent2, sent3, sent4, sent5, sent6, sent7]

  def prompt(self):
    story = " ".join(self.premise())
    return story

  def question(self):
    if self.Q == 'reality':
      return 'What was in the {}?'.format(self.place)
    elif self.Q == 'belief':
      return 'What was supposed to be in the {}?'.format(self.place)
    elif self.Q =='1stA':
      return "After {} opened the {}, what would {} expect to find in the {}?".format(self.name[1], self.place, self.name[0], self.place)
    elif self.Q == '1stB':
      return "After {} opened the {}, what would {} expect to find in the {}?".format(self.name[1], self.place, self.name[1], self.place)
    elif self.Q == '2ndA':
      return "After {} opened the {}, what would {} think {} would expect to find in the {}?".format(self.name[1], self.place,  self.name[0], self.name[1], self.place)
    elif self.Q == '2ndB':
      return "After {} opened the {}, what would {} think {} would expect to find in the {}?".format(self.name[1], self.place, self.name[1], self.name[0], self.place)
  def answer(self):
    if self.Q == 'reality':
      return '{}'.format(self.obj[1])
    elif self.Q == 'belief':
      return '{}'.format(self.obj[0])
    elif self.Q == '1stA':
      return "{}".format(self.obj[1])
    elif self.Q =='1stB':
      return "{}".format(self.obj[0])
    elif self.Q == '2ndA':
      return "{}".format(self.obj[0])
    elif self.Q == '2ndB':
      return "{}".format(self.obj[0])
  def qa_prompt(self):
    start = 'Answer the question based on context:\n\nContext:'
    middle = "\nQuestion:\n"
    output = start + self.prompt() + middle + self.question()
    return output
  def mc_prompt(self):
    start = "Choose the correct answer from A or B for the following question:\nQuestion:\n"
    end = '\nA. {}\nB. {}'.format(self.obj[1], self.obj[0])
    output = start + self.prompt() + " " + self.question() + end
    return output
  def comp_prompt(self):
    start = "Complete the following paragraph:\n"
    if self.Q == 'reality':
      end = "In the {}, there was ".format(self.place)
    elif self.Q == 'belief':
      end = "In the {}, there was supposed to be ".format(self.place)
    elif self.Q =='1stA':
      end = "After {} opened the {}, {} would expect to find".format(self.name[1], self.place, self.name[0])
    elif self.Q == '1stB':
      end = "After {} opened the {}, {} would expect to find".format(self.name[1], self.place, self.name[1])
    elif self.Q == '2ndA':
      end = "After {} opened the {}, {} would think {} would expect to find".format(self.name[1], self.place, self.name[0], self.name[1])
    elif self.Q == '2ndB':
      end = "After {} opened the {}, {} would think {} would expect to find".format(self.name[1], self.place, self.name[1], self.name[0])
    output = start + self.prompt() + " " + end
    return output
  def fb_prompt(self):
    start = "Fill in the blank (< >):\n"
    middle = self.comp_prompt().split("\n")[-1].split(' ')
    middle[-1] = middle[-1] + " a <>."
    output = start + " ".join(middle)
    return output
  def tf_prompt(self):
    start = "Given the context, judge True or False of the given statements A and B respectively:\nContext:\n"
    middle = self.prompt() + " " + self.question() + "\nStatement:\n"
    if self.Q == "reality":
      statementA = "A. In the {}, there was a {}.\n".format(self.place, self.obj[1])
      statementB = "B. In the {}, there was a {}.\n".format(self.place, self.obj[0])
    elif self.Q == 'belief':
      statementA = "A. In the {}, there was supposed to be a {}.\n".format(self.place, self.obj[1])
      statementB = "B. In the {}, there was supposed to be a {}.\n".format(self.place, self.obj[0])
    elif self.Q =='1stA':
      statementA = "A. After {} opened the {}, {} would expect to find a {}.\n".format(self.name[1], self.place, self.name[0], self.obj[1])
      statementB = "B. After {} opened the {}, {} would expect to find a {}.\n".format(self.name[1], self.place,self.name[0], self.obj[0])
    elif self.Q == '1stB':
      statementA = "A. After {} opened the {}, {} would expect to find a {}.\n".format(self.name[1], self.place, self.name[1], self.obj[1])
      statementB = "B. After {} opened the {}, {} would expect to find a {}.\n".format(self.name[1], self.place, self.name[1], self.obj[0])
    elif self.Q == '2ndA':
      statementA = "A. After {} opened the {}, {} would think {} would expect to find a {}.\n".format(self.name[1], self.place, self.name[0], self.name[1], self.obj[1])
      statementB = "B. After {} opened the {}, {} would think {} would expect to find a {}.\n".format(self.name[1], self.place, self.name[0], self.name[1], self.obj[0])
    elif self.Q == '2ndB':
      statementA = "A. After {} opened the {}, {} would think {} would expect to find a {}.\n".format(self.name[1], self.place, self.name[1], self.name[0], self.obj[1])
      statementB = "B. After {} opened the {}, {} would think {} would expect to find a {}.\n".format(self.name[1], self.place, self.name[1], self.name[0], self.obj[0])
    end = "Use this format for the answer:\nA.\nB."
    output = start + middle + statementA  + statementB + end
    return output
  def tfr_prompt(self):
    start = "Given the context, reason through statements A and B step by step and provide a True or False judgment based on the reasoning:"
    output = "\n".join([start] + self.tf_prompt().split('\n')[1:])
    return output
```

## Generate stories

```python
output_list_smarties = []
random.seed(42)
name_list = random.choices(list(permutations(name, 2)), k = 30)
object_list = random.choices(list(permutations(objects, 2)), k = 30)
for i in range(0, 30):
  n = name_list[i]
  p = place[i%len(place)]
  o = object_list[i]
  l = location[i%len(location)]
  for que in ['reality', 'belief', '1stA', '1stB', '2ndA', '2ndB']:
    smarties = data_smarties(n, o, l, p, que)
    output_list_smarties.append([smarties.prompt(), smarties.question(), smarties.answer(), que, i+1,
                                 smarties.qa_prompt(), smarties.comp_prompt(), smarties.fb_prompt(), smarties.mc_prompt(),
                                 smarties.tf_prompt(), smarties.tfr_prompt()])
df_smarties = pd.DataFrame(output_list_smarties, columns = ['prompt', 'question', 'short_answer', 'question_type', 'story_index',
                                                            'qa_prompt', 'comp_prompt', 'fb_prompt', 'mc_prompt', 'tf_prompt', 'tfr_prompt'])
```
